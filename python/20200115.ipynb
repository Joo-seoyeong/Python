```python
movieIdDf=pd.DataFrame()
movieIdDf_temp=pd.DataFrame()
movieIdDf_temp=movieIdDf_temp.append({"movieCd":"","movieNm":"","movieNmEn":"","openDt":"", "directorpeopleNm":""}, ignore_index=True)
movieIdDf_temp
for i in range(0,len(soup.find_all("moviecd"))):
    movieIdDf_temp.at[0,"movieCd"]=soup.find_all("moviecd")[i].text
    movieIdDf_temp.at[0,"movieNm"]=soup.find_all("movienm")[i].text
    movieIdDf_temp.at[0,"movieNmEn"]=soup.find_all("movienmen")[i].text
    movieIdDf_temp.at[0,"openDt"]=soup.find_all("opendt")[i].text
    
    if soup.find_all("directors")[i].find_all("peoplenm") != []:
        movieIdDf_temp.iat[0,movieIdDf_temp.columns.get_loc("directorpeopleNm")]=soup.find_all("directors")[i].find_all("peoplenm")[0].text
    else:
        movieIdDf_temp.iat[0,movieIdDf_temp.columns.get_loc("directorpeopleNm")]=""
    
    movieIdDf=movieIdDf.append(movieIdDf_temp, ignore_index=True)
    
    for j in range(0,len(movieIdDf_temp.columns)):
        movieIdDf_temp.iat[0,j]=""
    
movieIdDf
```
