##### 위와 같은 데이터를 가지고 stock_price_DF_withIndex에서 "Gap"이라는 컬럼을 생성하고 이 컬럼에서는 시가/종가/고가/저가 중에서 가장 큰 값과 가장 작은 값의 차이를 기록하는 컬럼을 생성

```python
stock_price_DF_withIndex["Gap"]=stock_price_DF_withIndex.apply(lambda x : x.max()-x.min(), axis=1)
stock_price_DF_withIndex
```

```python
stock_price_DF_withIndex.iloc[0,0:4]
```

```python
stock_price_DF_withIndex["Gap_forloop"]=0
labels=["Gap_forloop"]
index_ints=[stock_price_DF_withIndex.columns.get_loc(label) for label in labels]

for i in range(0,len(stock_price_DF_withIndex)):
    temp=stock_price_DF_withIndex.iloc[i,0:4].max()-stock_price_DF_withIndex.iloc[i,0:4].min()
    stock_price_DF_withIndex.iloc[i,index_ints[0]]=temp
stock_price_DF_withIndex
```

```python
movieIdDF=pd.DataFrame()
movieIdDF_temp=pd.DataFrame()
movieIdDF_temp=movieIdDF_temp.append({"movieCd":"","movieNm":"","movieNmEn":"","openDt":"", "directorpeopleNm":""}, ignore_index=True)
movieIdDF_temp
for i in range(0,40):
    # 데이터 담기
    movieIdDF_temp.at[0,"movieCd"]=movieIdData["movieListResult"]["movieList"][i]["movieCd"]
    movieIdDF_temp.at[0,"movieNm"]=movieIdData["movieListResult"]["movieList"][i]["movieNm"]
    movieIdDF_temp.at[0,"movieNmEn"]=movieIdData["movieListResult"]["movieList"][i]["movieNmEn"]
    movieIdDF_temp.at[0,"openDt"]=movieIdData["movieListResult"]["movieList"][i]["openDt"]
    
    if movieIdData["movieListResult"]["movieList"][i]["directors"]!=[]:
        movieIdDF_temp.iat[0,movieIdDF_temp.columns.get_loc("directorpeopleNm")]=movieIdData["movieListResult"]["movieList"][i]["directors"][0]["peopleNm"]
    else:
        movieIdDF_temp.iat[0,movieIdDF_temp.columns.get_loc("directorpeopleNm")]=""
    # 합치기
    movieIdDF = movieIdDF.append(movieIdDF_temp, ignore_index=True)
    
    # 비우기
    for j in range(0,len(movieIdDF_temp.columns)):
        movieIdDF_temp.iat[0,j]=""


movieIdDF.head()
```
