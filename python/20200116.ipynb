```python
# 01) DataSet
baseball_player=pd.DataFrame(
    {
    "player_Name":
        ["권혁","이범호","필","박병호","양준혁"],
    "team_No":
        [1993,2001,2001,2010,None]
    }
)   

baseball_team=pd.DataFrame(
    {
        "team_Name":["한화","두산","기아","넥센"],
        "team_Home":["대전","서울","광주","고척"],
        "team_id":[1993,1999,2001,2010]
    }
)
```

```python
baseball_player
```

```python
baseball_team
```

```python
# left outter join
pd.merge(baseball_player,baseball_team,how='left',left_on="team_No",right_on="team_id")
```

```python
# inner join
pd.merge(baseball_player, baseball_team,how='inner',left_on="team_No",right_on="team_id")
```

```python
# right outter
pd.merge(baseball_player, baseball_team, how='right', left_on="team_No",right_on="team_id")
```

```python
# left excluding join
# 01) left join
temp=pd.merge(baseball_player, baseball_team, how='left',left_on="team_No",right_on="team_id")
# 02) select NaN
temp[temp["team_id"].isnull()]
```

```python
# right excluding join
# 01) right join
temp=pd.merge(baseball_player, baseball_team, how='right',left_on="team_No",right_on="team_id")
# 02) select NaN
temp[temp["player_Name"].isnull()]
```

```python
# union
pd.merge(baseball_player, baseball_team, how='outer',left_on="team_No",right_on="team_id")
```

```python
# full outer excluding join
temp=pd.merge(baseball_player, baseball_team,how='outer',left_on="team_No",right_on="team_id")
temp[temp["player_Name"].isnull() | temp["team_Name"].isnull()]
```

### Pivot

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("C:\\Users\\1-10\\Desktop\\python\\sampledata\\11_sales-funnel.xlsx")
data.head()
```

```python
#  좀 더 명확하게 하기위해서 카테고리 변수화 - 벡터라이징
data["Status"]=data["Status"].astype("category")

data["Status"].cat.set_categories(["won","pending","presented","declined"],
    inplace=True)

data.head()
```

```python
# Name을 가로줄 기준으로 하고 데이터를 보자 -> 숫자인 컬럼들만 들어온다
pd.pivot_table(data,index=["Name"]) # index에 여러 개가 들어올 수 있으니까 리스트로 받는다
```

```python
# 가로줄 여러 계층 구조로 사용할 경우
pd.pivot_table(data, index=["Name","Rep","Manager"])
```

```python
# 2개만 가로줄 선택
pd.pivot_table(data, index=["Manager","Rep"])
```

```python
# row - manager / rep,column - price
pd.pivot_table(data, index=["Manager","Rep"], values=["Price"])
```

```python
# pivot with aggfunc : sum
import numpy as np
pd.pivot_table(data, index=["Manager","Rep"],
               values=["Price"],
               aggfunc=np.sum)
```

```python
# pivot with aggfunc -> make new column
pd.pivot_table(data, index=["Manager","Rep"],
              values=["Price"],
              aggfunc=[np.mean, len]) # 인수로 정의된 함수를 수행해서 대표값 계산
```

```python
# 매니저 별로 담당하는 직원들의 실적을 보고싶을 때 
# 가로-매니저/담당자  세로-실적(평균이 아니라 전체 판매한 실적 총합)
# 세로줄에 대한 계층
pd.pivot_table(data, index=["Manager","Rep"],
              values=["Price"],
              columns=["Product"],
              aggfunc=[np.sum])
```

```python
pd.pivot_table(data, index=["Manager","Rep"],
              values=["Price"],
              columns=["Product"],
              aggfunc=[np.sum],
              fill_value=0) # 빈 값에 대해 0으로 채움
```

```python
# 여러 컬럼에 여러 계층 구조
pd.pivot_table(data, index=["Manager","Rep"],
              values=["Price","Quantity"],
              columns=["Product"],
              aggfunc=[np.sum],
              fill_value=0) 
```

```python
# 매니저 별로 담당하는 직원들의 실적을 보고싶을 때
# 품목에 대한 정보를 가로 인덱스로
pd.pivot_table(data, index=["Manager","Rep","Product"],
              values=["Price","Quantity"],
              aggfunc=[np.sum],
              fill_value=0)
```

```python
# Total Option
pd.pivot_table(data, index=["Manager","Rep","Product"],
              values=["Price","Quantity"],
              aggfunc=[np.sum,np.mean],
              fill_value=0,
              margins=True)
```

```python
# row : manager/status
# column : Quantity[len] / Price[sum]
# column category:product
pd.pivot_table(data, index=["Manager","Status"],
              columns=["Product"],
              values=["Quantity","Price"],
              aggfunc={"Quantity":len,"Price":np.sum}, # 딕셔너리를 활용해서 각각의 컬럼에 대해서 다른 함수를 지정
              fill_value=0) 
```

```python
# aggfunc에서도 계층 구조 적용
pd.pivot_table(data, index=["Manager","Status"],
              columns=["Product"],
              values=["Quantity","Price"],
              aggfunc={"Quantity":len,"Price":[np.sum,np.mean]},
              fill_value=0)
```

```python
table=pd.pivot_table(data, index=["Manager","Status"],columns=["Product"],values=["Quantity","Price"],
                    aggfunc={"Quantity":len, "Price":[np.sum,np.mean]},
                    fill_value=0)
```

```python
# crosstab : 빈도에 초점을 맞추고 있음
pd.crosstab(data["Name"],data["Price"])
```

```python
# 여러 줄을 선택하기 위해서는 리스트 사용
pd.crosstab([data["Manager"],data["Rep"]], [data["Price"],data["Quantity"]])
```
